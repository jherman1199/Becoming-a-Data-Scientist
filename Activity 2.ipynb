### Activity 2 - Becoming a Data Scientist - Creating Visuals for Exploratory Data Analysis

Dataset I chose is on Pima Indians with Diabetes.  Dataset can be found here: https://archive.ics.uci.edu/ml/datasets/Pima+Indians+Diabetes

```python
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
matplotlib.style.use('fivethirtyeight')
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('pima_indians_diabetes.csv')
```

```python
df.head()
```

```python
df['plasma_glucose_concentration'] = df['plasma_glucose_concentration'].replace('0', np.nan)
df['diastolic_blood_pressure'] = df['diastolic_blood_pressure'].replace('0', np.nan) 
df['tricepts_skin_fold_thickness'] = df['tricepts_skin_fold_thickness'].replace('0', np.nan) 
df['2_hour_serum_insulin'] = df['2_hour_serum_insulin'].replace('0', np.nan)        
df['body_mass_index'] = df['body_mass_index'].replace('0', np.nan) 
df['diabetes_pedigree_function'] = df['diabetes_pedigree_function'].replace('0', np.nan) 
df['age'] = df['age'].replace('0', np.nan) 
```

```python
df.isnull().sum()
```

Due to the high number of missing values in tricepts_skin_fold_thickness and 2_hour_serum_insulin I will drop them while doing my exploratory visualization.  For the other values, I will add the median value for the NaN values

## Age vs Diabetes

```python
figure = plt.figure(figsize=(15,8))
plt.hist([df[df['class_variable']==1]['age'],df[df['class_variable']==0]['age']], stacked=True, color = ['g','r'],
         bins = 30,label = ['Diabetes','No-Diabetes'])
plt.xlabel('Age')
plt.ylabel('Number of Pima Indians')
plt.legend()
```

## Times Pregnant vs Diabetes

```python
figure = plt.figure(figsize=(15,8))
plt.hist([df[df['class_variable']==1]['times_pregnant'], df[df['class_variable']==0]['times_pregnant']], stacked=True,
         color = ['g','r'],bins = 50,label = ['Diabetes','No-Diabetes'])
plt.xlabel('Number of Times Pregnant')
plt.ylabel('Number of Pima Indians')
plt.legend()
```

```python
df['body_mass_index'].fillna(df['body_mass_index'].median(), inplace=True)
```

## Body Mass Index vs Diabetes

```python
figure = plt.figure(figsize=(15,8))
plt.hist([df[df['class_variable']==1]['body_mass_index'], df[df['class_variable']==0]['body_mass_index']], stacked=True,
         color = ['g','r'],bins = 50,label = ['Diabetes','No-Diabetes'])
plt.xlabel('Body Mass Index')
plt.ylabel('Number of Pima Indians')
plt.legend()
```

```python
df['plasma_glucose_concentration'].fillna(df['plasma_glucose_concentration'].median(), inplace=True)
```

## Plasma Glucose Concentration vs Diabetes

```python
figure = plt.figure(figsize=(15,8))
plt.hist([df[df['class_variable']==1]['plasma_glucose_concentration'], 
          df[df['class_variable']==0]['plasma_glucose_concentration']], stacked=True,
         color = ['g','r'],bins = 50,label = ['Diabetes','No-Diabetes'])
plt.xlabel('Plasma Glucose Concentration')
plt.ylabel('Number of Pima Indians')
plt.legend()
```

```python
df['diastolic_blood_pressure'].fillna(df['diastolic_blood_pressure'].median(), inplace=True)
```

## Diastolic Blood Pressure vs Diabetes

```python
figure = plt.figure(figsize=(15,8))
plt.hist([df[df['class_variable']==1]['diastolic_blood_pressure'], 
          df[df['class_variable']==0]['diastolic_blood_pressure']], stacked=True,
         color = ['g','r'],bins = 50,label = ['Diabetes','No-Diabetes'])
plt.xlabel('Diastolic Blood Pressure')
plt.ylabel('Number of Pima Indians')
plt.legend()
```
